```python
# 아래는 카이제곱 코드입니당 -----------------------------------------------------------------------------------
import scipy.stats as stats

df_merge2=pd.merge(df_broad,df_crew,on='상품ID',how='left')
df_merge=pd.merge(df_merge2,df_product,on="상품ID",how='left')
df_broad["방송Code"] = df_crew["방송Code"]
df_bandc=pd.merge(df_broad,df_crew,on = '방송Code',how = "inner")
df_bandc['목표달성여부'] = df_bandc['상품순주문금액'] - df_bandc['상품목표주문금액']
```

```python
# 목표변수 안될시 이 함수와 아래 apply 코드를 쓰세용
def conversion(x):
  if x>=0 : x=1
  else : x=0
  return x
```

```python
df_bandc['목표달성여부']=df_bandc['목표달성여부'].apply(lambda x : conversion(x))
```

```python
df_merge['시간대'] = df_merge['방송시작시간'].apply(lambda x : str(x)[-5:-3])
df_bandc=pd.merge(df_bandc,df_product[['카테고리','상품ID']],left_on='상품ID_x',right_on='상품ID',how = "left")
```

```python
df = pd.crosstab(df_merge["시간대"], df_bandc['목표달성여부'])
chi, pval, dof, expected = stats.chi2_contingency(df)
print("chi-square test")
print("chisq: {0:0.3f}".format(chi))
print("p: {0:0.3f}".format(pval))
df
# 시간대는 목표달성여부와 관련성이 있음
```
