```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
```

```python
value = data[(data['목표달성여부'] == 1)]
```

```python
value
```

```python
def conversion(x):
  if x>=0 : x='성공'
  else : x='실패'
  return x
```

```python
value['목표달성여부']=value['목표달성여부'].apply(lambda x : conversion(x))
```

```python
import datetime
value['hour'] = value['방송시작시간'].dt.hour
```

```python
value['weekday'] = value['방송시작시간'].dt.weekday

value['weekday'].replace({0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}, inplace = True)
```

```python
value['month'] = value['방송시작시간'].dt.month
```

```python
value['month'].replace(1, '겨울', inplace = True)
value['month'].replace(2, '겨울', inplace = True)
value['month'].replace(3, '봄', inplace = True)
value['month'].replace(4, '봄', inplace = True)
value['month'].replace(5, '봄', inplace = True)
value['month'].replace(6, '여름', inplace = True)
value['month'].replace(7, '여름', inplace = True)
value['month'].replace(8, '여름', inplace = True)
value['month'].replace(9, '가을', inplace = True)
value['month'].replace(10, '가을', inplace = True)
value['month'].replace(11, '가을', inplace = True)
value['month'].replace(12, '겨울', inplace = True)
```

```python
value.info()
```

```python
value['hour'] = value['hour'].astype('str')
```

```python
value.info()
```

```python
value.reset_index(drop=True, inplace = True)
```

```python
value['카테고리'].value_counts()
```

```python
t=[]
for i in range(len(value.hour)):
    if value.목표달성여부[i] == '성공':
        t.append([value.hour[i],value.weekday[i],value.month[i]])
print(type(t))    
```

```python
t[0][0]
```

```python
encoding = pd.DataFrame(t)
```

```python
enc = TransactionEncoder()
df_raw_enc=enc.fit_transform(X=t)
df_asso = pd.DataFrame(df_raw_enc,columns=enc.columns_)
df_asso.head()
```

```python
print(type(t[0][0]))
```

```python
v_min_support=0.005
df_freq = apriori(df_asso,min_support=v_min_support,use_colnames=True)
df_asso_rule_시간 = association_rules(df_freq,metric = 'confidence',min_threshold=0.005)
```

```python
df_asso_rule_시간
df_asso_rule_시간.sort_values(by=df_asso_rule_시간.columns[6], axis=0,ascending=False).head(10)
```

```python
def my_draw_nx(asso_rule_df, title = 'Network Graph', fontsize = 15, show_edgelabel = True):
    import networkx as nx
    
    nx_df = asso_rule_df.copy()
    nx_df['antecedents'] = nx_df['antecedents'].apply(lambda x : str(set(x)))
    nx_df['consequents'] = nx_df['consequents'].apply(lambda x : str(set(x)))
    
    g = nx.from_pandas_edgelist(nx_df, source = 'antecedents',
                               target = 'consequents',
                               edge_attr = ['support', 'confidence', 'lift'])
    print(nx.info(g))
    
    d1 = {row.antecedents : row['antecedent support']
         for ix, row in nx_df.iterrows()}
    d2 = {row.consequents : row['consequent support']
         for ix, row in nx_df.iterrows()}
    d1.update(d2)
    node_size = np.array([np.sqrt(d1[k]) * 600 for k in list(g)])
    
    plt.figure(figsize = (10, 10))
    pos = nx.spring_layout(g, weight = 'confidence', seed = 42)
    nx.draw_networkx_labels(g, pos, font_size = fontsize, )
    nx.draw_networkx(g, pos, node_color = 'green', with_labels = False, alpha = 0.3, arrows = True, arrowsize = fontsize,
                    node_size = node_size)
    if show_edgelabel:
        edge_d = { (row.antecedents, row.consequents) :f'{row.lift:.2f}'
                 for ix, row in nx_df.iterrows()}
        nx.draw_networkx_edge_labels(g, pos, edge_labels = edge_d)
        
    plt.title(title)
    plt.show()
```

```python
my_draw_nx(df_asso_rule_시간, title = 'time')
```
