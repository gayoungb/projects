```python
value = data[(data['목표달성여부'] == 1)]
value.head()
```

```python
def conversion(x):
  if x>=0 : x='성공'
  else : x='실패'
  return x
```

```python
value['목표달성여부']=value['목표달성여부'].apply(lambda x : conversion(x))
```

```python
import datetime
value['hour'] = value['방송시작시간'].dt.hour
value['weekday'] = value['방송시작시간'].dt.weekday
value['weekday'].replace({0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}, inplace = True)
value['month'] = value['방송시작시간'].dt.month
value['season'] = value['month']
value['season'] = value['season'].replace(1, '겨울')
value['season'] = value['season'].replace(2, '겨울')
value['season'] = value['season'].replace(3, '봄')
value['season'] = value['season'].replace(4, '봄')
value['season'] = value['season'].replace(5, '봄')
value['season'] = value['season'].replace(6, '여름')
value['season'] = value['season'].replace(7, '여름')
value['season'] = value['season'].replace(8, '여름')
value['season'] = value['season'].replace(9, '가을')
value['season'] = value['season'].replace(10, '가을')
value['season'] = value['season'].replace(11, '가을')
value['season'] = value['season'].replace(12, '겨울')
value['hour'] = value['hour'].astype('str')
value['month'].replace(1, 'jan', inplace = True)
value['month'].replace(2, 'feb', inplace = True)
value['month'].replace(3, 'mar', inplace = True)
value['month'].replace(4, 'apr', inplace = True)
value['month'].replace(5, 'may', inplace = True)
value['month'].replace(6, 'jun', inplace = True)
value['month'].replace(7, 'jul', inplace = True)
value['month'].replace(8, 'aug', inplace = True)
value['month'].replace(9, 'sep', inplace = True)
value['month'].replace(10, 'oct', inplace = True)
value['month'].replace(11, 'nov', inplace = True)
value['month'].replace(12, 'dec', inplace = True)
value['month'] = value['month'].astype('str')
```

## 카테고리별 연관규칙

### 신선농산

```python
c=[]
for i in range(len(value.hour)):
    if value.카테고리[i] == '신선농산':
        c.append([value.hour[i],value.weekday[i],value.month[i],value.season[i]])
print(type(c))

encoding = pd.DataFrame(c)

enc = TransactionEncoder()
df_raw_enc=enc.fit_transform(X=c)
df_asso = pd.DataFrame(df_raw_enc,columns=enc.columns_)
df_asso.head()

v_min_support=0.005
df_freq = apriori(df_asso,min_support=v_min_support,use_colnames=True)
df_asso_rule_신선 = association_rules(df_freq,metric = 'confidence',min_threshold=0.005)
cond1 = (df_asso_rule_신선['support'] > 0.01)
df_asso_rule_신선.loc[cond1].sort_values(by=df_asso_rule_신선.columns[6], axis=0,ascending=False).head(10)
```

```python
data = pd.read_csv('total_data.csv')
```

```python
data['방송시작시간'] = data['방송시작시간'].astype('datetime64')
```

```python
def conversion(x):
  if x==1 : x='성공'
  else : x='실패'
  return x
```

```python
data['목표달성여부']=data['목표달성여부'].apply(lambda x : conversion(x))
```

```python
import datetime
data['hour'] = data['방송시작시간'].dt.hour
data['weekday'] = data['방송시작시간'].dt.weekday
data['weekday'].replace({0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}, inplace = True)
data['month'] = data['방송시작시간'].dt.month
data['month'].replace(1, '겨울', inplace = True)
data['month'].replace(2, '겨울', inplace = True)
data['month'].replace(3, '봄', inplace = True)
data['month'].replace(4, '봄', inplace = True)
data['month'].replace(5, '봄', inplace = True)
data['month'].replace(6, '여름', inplace = True)
data['month'].replace(7, '여름', inplace = True)
data['month'].replace(8, '여름', inplace = True)
data['month'].replace(9, '가을', inplace = True)
data['month'].replace(10, '가을', inplace = True)
data['month'].replace(11, '가을', inplace = True)
data['month'].replace(12, '겨울', inplace = True)
data['hour'] = data['hour'].astype('str')
```

```python
df = pd.crosstab(data['hour'],data['목표달성여부'])
chi, pval, dof, expected = chi2_contingency(df)
print('chi-squeare test')
print('chisq: {0:0.3f}'.format(chi))
print('p: {0:0.3f}'.format(pval))
df
```
